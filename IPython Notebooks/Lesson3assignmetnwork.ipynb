```python
import math
import numpy
from matplotlib import pyplot
```

```python
N = 51
x_start, x_end = -1.0, 2.0
y_start, y_end = -0.5, 0.5
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X,Y = numpy.meshgrid(x,y)
```

```python
NACA_x = numpy.loadtxt('resources\\NACA0012_x.txt')
```

```python
NACA_y = numpy.loadtxt('resources\\NACA0012_y.txt')
```

```python
NACA_strength = numpy.loadtxt('resources\\NACA0012_sigma.txt')
```

```python
def get_velocity_NACA(strength,xs,ys,X,Y):
    """Returns the velocity field generated for this NACA
    
    Arguments
    ----------
    strength -- strength of naca
    xs,ys -- coordinates of naca
    X,Y, mesh grid.
    """
    
    u = strength/(2*numpy.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*numpy.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    
    return u,v
```

```python
def get_stream_function(strength, xs, ys, X, Y):
    """Returns the stream-function generated by a source/sink.
    
    Argumetns
    ------
    strength -- strength of the source/sink
    xs, ys -- coordinates of the source/sink
    X,Y -- mesh grid
    """
    
    psi = strength/(2*numpy.pi)*numpy.arctan2((Y-ys), (X-xs))
    
    return psi
```

```python
u_inf = 1.0
u_freestream = u_inf * numpy.ones((N,N), dtype=float)
v_freestream = numpy.zeros((N,N),dtype=float)
psi_freestream = u_inf*Y
```

```python
u = u_freestream
v = v_freestream
psi = psi_freestream

for i in range(100):
    u_NACA, v_NACA = get_velocity_NACA(NACA_strength[i], NACA_x[i], NACA_y[i], X,Y)
    psi_NACA = get_stream_function(NACA_strength[i], NACA_x[i], NACA_y[i], X,Y)
    u = u+u_NACA
    v = v+v_NACA
    psi = psi+psi_NACA
    
```

```python

%matplotlib inline

size=40
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u,v,density=2,linewidth=1,arrowsize=1,arrowstyle='->')
pyplot.scatter(NACA_x, NACA_y, color='#CD2305', s=80, marker='o')

x_stag = NACA_x - NACA_strength/(2*numpy.pi*u_inf)
y_stag = NACA_y

pyplot.scatter(x_stag, y_stag, color='g',s=80,marker='o')

#pyplot.contour(X,Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid');
```

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

#linindex
maxloc = numpy.argmax(cp)
a = numpy.unravel_index(maxloc, cp.shape)

# plots the pressure coefficient field
size = 40
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-1.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-1.0, -0.5, 0.0, 0.5, 1.0])
pyplot.scatter(NACA_x, NACA_y, color='#CD2305', s=80, marker='o')
#pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter(X[a], Y[a], color='#CD2305', s=200, marker='o');
```

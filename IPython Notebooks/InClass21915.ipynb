```python
N = 100
x_start, x_end = -5.0, 5.0 #-2.0, 2.0
y_start, y_end = -5.0, 5.0 #-0.2, 0.4
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)

theta = numpy.linspace(0, 2*numpy.pi,N)
r = 1.15
c=1.0
centerx = -0.15
centery = 0.0
```

```python
def transformation(Es, Ns):
    """This uses the transform propertiesnessessnessess
    
    Argumentssssssss
    ----------------
    Es, Ns -- transformation properties
    X, Y -- mesh grid
    """
    
    x = ((Es)*((Es**2) + (Ns**2) + 1))/((Es**2)+(Ns**2))
    y = ((Ns)*((Es**2) + (Ns**2) - 1))/((Es**2)+(Ns**2))
    
    return x, y
```

```python
def transformation_stream(Es, Ns, X, Y):
    """This uses the transform propertiesnessessnessess
    
    Argumentssssssss
    ----------------
    Es, Ns -- transformation properties
    X, Y -- mesh grid
    """
    
    x = ((Es-X)*(((Es-X)**2) + ((Ns-Y)**2) + 1))/(((Es-X)**2)+((Ns-Y)**2))
    y = ((Ns-Y)*(((Es-X)**2) + ((Ns-Y)**2) - 1))/(((Es-X)**2)+((Ns-Y)**2))
    
    return x, y
```

```python
def circle (theta, r, xc, yc):
    """circles
    
    ag-------
    xc, yc -- cord
    X,Y, -- grid
    r -- radius
    """
    x = r*numpy.cos(theta) + xc
    y = r*numpy.sin(theta) + yc
    
    return x, y
```

```python
#define circle first
cx, cy = circle(theta, r, centerx, centery)

#plug into transform function
#compute
xt, yt = transformation(cx, cy)
xts, yts = transformation_stream(cx, cy, X, Y)
```

```python
def get_stream_fn(u_inf, alpha, X, Y, Nx, Ny):
    """Returns the stream-functions
    
    Arguments
    ---------
    u_inf -- strength of stream flow
    alpha -- angle of attack
    X,Y -- mesh grid
    Nx, Ny -- number of points in x and y.
    """
    u = u_inf*numpy.cos(alpha) * numpy.ones((Nx, Ny))
    v = u_inf*numpy.sin(alpha) * numpy.ones((Nx, Ny))
    psi = u_inf*(Y*numpy.cos(alpha) - X*numpy.sin(alpha))
    
    return u, v, psi
   
```

```python
def get_doublet(strength,xd,yd,X,Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet
    xd, yd -- coordinates of the doublet
    X,Y -- mesh grid.
    """
    
    u = -strength/(2*numpy.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = -strength/(2*numpy.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    psi = -strength/(2*numpy.pi)*((Y-yd)/((X-xd)**2+(Y-yd)**2)**2)  
    
    return u, v, psi
```

```python
def get_vortex(strength, xv, yv, X, Y):
    """Returns the velocity field generated by a vortex.
    
    Arguments
    ---------
    strength -- strength of the vortex.
    xv, yv -- coordinates of the vortex.
    X, Y -- mesh grid.
    """
    
    u = + strength/(2*numpy.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*numpy.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    psi = strength/(4*numpy.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return u, v, psi
```

```python
import math
import numpy
from matplotlib import pyplot
```

```python
N = 100
x_start, x_end = 2.0, -2.0
y_start, y_end = 0.4, -0.4
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    """Return the velocity of a generated vortex.
    
    Arguments
    ----------
    strength = strength of the vortex
    xv, yv = coordinates of vortex
    X, Y = mesh grid.
    """
    
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v

def get_velocity_vortex_inf(strength, xv, yv, a, X, Y):
    """Returns the velocity of a generated vortex for infinite vortices.
    
    Arguments
    ---------
    strength = strength of the vortex
    xv, yv = coordinates of vortex
    X, Y = mesh grid.
    """
    
    u = + strength/(2*a)*((numpy.sinh((2*math.pi*Y-yv)/(a)))/((numpy.cosh((2*math.pi*Y-yv)/(a)))-(numpy.cos((2*math.pi*X-xv)/(a)))))
    v = - strength/(2*a)*((numpy.sin((2*math.pi*X-xv)/(a)))/((numpy.cosh((2*math.pi*Y-yv)/(a)))-(numpy.cos((2*math.pi*X-xv)/(a)))))
    
    return u, v
    
```

```python
u_inf = 0.0
u_freestream = u_inf * numpy.ones((N,N), dtype=float)
v_freestream = numpy.zeros((N,N),dtype=float)
psi_freestream = u_inf*Y

u = u_freestream
v = v_freestream
psi = psi_freestream
V_strength = 10
a = 0.5
V_x = -N/4
V_y = 0.0

for i in range(N):
    u_V, v_V = get_velocity_vortex(V_strength, V_x, V_y, X,Y)
    u = u+u_V
    v = v+v_V
    V_x = V_x + a
 
    
%matplotlib inline

size=40
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X,Y,u,v,density=0.8,linewidth=3,arrowsize=5,arrowstyle='->')

for i in range(N):
    pyplot.scatter(V_x, V_y, color='#CD2305', s=1000, marker='o');
    V_x = V_x - a
```

```python
u2_freestream = u_inf * numpy.ones((N,N), dtype=float)
v2_freestream = numpy.zeros((N,N),dtype=float)

u2 = u2_freestream
v2 = v2_freestream
V2_strength = 50
a2 = 0.5
V2_x = 0.0
V2_y = 0.0

u_V2, v_V2 = get_velocity_vortex_inf(V2_strength, V2_x, V2_y, a2, X,Y)
    
%matplotlib inline

size=40
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X,Y,u_V2,v_V2,density=0.8,linewidth=3,arrowsize=5,arrowstyle='->')

NN = 9
Vx_start, Vx_end = 2.0, -2.0
Vy_start, Vy_end = 0.4, -0.4
Vx = numpy.linspace(Vx_start, Vx_end, NN)
Vy = numpy.zeros_like(Vx)
pyplot.scatter(Vx, Vy, color='#CD2305', s=1000, marker='o');
```

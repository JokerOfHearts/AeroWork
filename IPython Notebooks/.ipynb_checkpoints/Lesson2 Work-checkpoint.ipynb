```python
from math import *
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
N = 200
```

```python
x_start, x_end = -4.0, 4.0
```

```python
y_start, y_end = -2.0, 2.0
```

```python
x = np.linspace(x_start, x_end, N)
```

```python
y = np.linspace(y_start, y_end, N)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
np.shape(X)
```

```python
np.shape(Y)
```

```python
u_inf = 1.0
```

```python
u_freestream = u_inf*np.ones((N,N),dtype=float)
v_freestream = u_inf*np.zeros((N,N),dtype=float)
psi_freestream = u_inf*Y
```

```python
def get_velocity(sigma, xs, ys, X, Y):
    """Returns the velocity field generated from a source/sink.
    
    Arguments
    ---------
    sigma -- strength of the source/sink
    xs, ys -- coordinates of the source/sink
    X,Y -- your mesh grid size
    """
    denom = ((X-xs)**2 + (Y-ys)**2)
    const_strength = sigma/(2*pi)
    
    u = const_strength*((X-xs)/(denom))
    v = const_strength*((Y-ys)/(denom))
    
    return u, v
```

```python
help (get_velocity)
```

```python
def get_stream_function(sigma, xs, ys, X, Y):
    """Returns the stream-functions generated by a source/sink
    
    Arguments
    ---------
    sigma -- strength of the source/sink
    xs, ys -- coordinates of the source/sink
    X,Y -- your mesh grid size
    """
    const_strength = sigma/(2*pi)
    
    psi = const_strength*np.arctan2((Y-ys),(X-xs))
    
    return psi
```

```python
sigma_source = 5.0
```

```python
x_source, y_source = -1.0, 0.0
```

```python
u_source, v_source = get_velocity(sigma_source, x_source, y_source, X, Y)
```

```python
psi_source = get_stream_function(sigma_source, x_source, y_source, X, Y)
```

```python
#superposition of the source on the freestream
u = u_freestream + u_source
v = v_freestream + v_source
psi = psi_freestream + psi_source

#plots the streamlines
%matplotlib inline

size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.grid(True)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X,Y,u,v,density=2,linewidth=1,arrowsize=1, arrowstyle='->')
plt.scatter(x_source,y_source,color='#CD2305',s=80,marker='o')

#calculates the stagpoint
x_stagnation = x_source - sigma_source/(2*pi*u_inf)
y_stagnation = y_source

#adding the stag point to fig
plt.scatter(x_stagnation,y_stagnation,color='g', s=80, marker='o')

#adds the dividing line to fig
plt.contour(X,Y,psi, levels = [-sigma_source/2,+sigma_source/2],colors='#CD2305',linewidth=2,linestyles='solid');
```

```python
sigma_sink = -5.0
x_sink, y_sink = 1.0, 0.0

#compute velocity field
u_sink, v_sink = get_velocity(sigma_sink, x_sink, y_sink, X,Y)

#computes stream function
psi_sink = get_stream_function(sigma_sink, x_sink, y_sink, X,Y)
```

```python
#super posotion of a source and a sink in a freestream
u=u_freestream+u_source+u_sink
v=v_freestream+v_source+v_sink
psi = psi_freestream+psi_source+psi_sink

#plots
size=10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.streamplot(X,Y,u,v,density=2,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter([x_source, x_sink],[y_source, y_sink],color='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi, levels=[0.],colors='#CD2305',linewidth=2,linestyles='solid');
```

```python
#computing the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

#plots the pressure coefficient field
size = 10
plt.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
contf= plt.contourf(X,Y,cp,levels=np.linspace(-2.0,1.0,100),extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter([x_source, x_sink],[y_source, y_sink], color='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,levels=[0.],colors='#CD2305', linewidth=2,linestyles='solid');
```

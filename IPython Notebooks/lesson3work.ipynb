```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#starting grid meshing
N=50
x_start,x_end =-2.0,2.0
y_start,y_end=-1.0,1.0
x=np.linspace(x_start,x_end,N)
y=np.linspace(y_start,y_end,N)
X,Y = np.meshgrid(x,y)
```

```python
kappa=1.0
x_doublet,y_doublet = 0.0,0.0
```

```python
def get_velocity_doublet(strength,xd,yd,X,Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet
    xd, yd -- coordinates of the doublet
    X,Y -- mesh grid.
    """
    
    u= -strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v= -strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v

def get_stream_fn_doublet(strength, xd, yd, X, Y):
    """Returns the stream-function generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet
    xd, yd -- coordinates of the doublet
    X,Y -- mesh grid.
    """
    psi = -strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
    
```

```python
#computes the velocity field
u_doublet, v_doublet = get_velocity_doublet(kappa, x_doublet, y_doublet, X,Y)

#compute the stream fn
psi_doublet = get_stream_fn_doublet(kappa, x_doublet,y_doublet,X,Y)
```

```python
%matplotlib inline

#time for plots
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X,Y,u_doublet,v_doublet,density=2,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(x_doublet,y_doublet,color='#CD2305',s=80,marker='o');
```

```python
#starting to work on uniform flow
u_inf = 1.0
u_freestream = u_inf*np.ones((N,N), dtype=float)
v_freestream = np.zeros((N,N), dtype=float)

psi_freestream = u_inf*Y
```

```python
#superposition of the doublet on the freestream folow
u=u_freestream+u_doublet
v=v_freestream+v_doublet
psi=psi_freestream+psi_doublet

#time to plot!
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X,Y,u,v,density=2,linewidth=1,arrowsize=1,arrowstyle='->')
plt.contour(X,Y,psi,levels=[0.],colors='#CD2305',linewidths=2,linestyle='solid')
plt.scatter(x_doublet,y_doublet,color='#CD2305',s=80,marker='o')

#Calc the stags
x_stag1, y_stag1 = +math.sqrt(kappa/(2*math.pi*u_inf)), 0
x_stag2, y_stag2 = -math.sqrt(kappa/(2*math.pi*u_inf)), 0

#Adding the stags to fig
plt.scatter([x_stag1,x_stag2],[y_stag1,y_stag2], color='g',s=80,marker='o');
```

```python
#computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

#plots the pressure coefficient field
size = 10
plt.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
contf = plt.contourf(X, Y, cp, levels=np.linspace(-2.0, 1.0, 100), extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')
plt.scatter([x_stag1, x_stag2], [y_stag1,y_stag2], color='g',s=80,marker='o')
plt.contour(X,Y,psi, levels=[0.], colors='#CD2305', linewidth=2,linestyles='solid');
```

```python
class Panel:
    """Contains information related to one panel."""
    def __init__(self, xa, ya, xb, yb):
        """Creates a panel.
        
        Arguments
        ---------
        xa, ya -- Cartesian coordinates of the first end-point.
        xb, yb -- Cartesian coordinates of the second end-point.
        """
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2       # control-point (center-point)
        self.length = math.sqrt((xb-xa)**2+(yb-ya)**2)# length of the panel
        
        # orientation of the panel (angle between x-axis and panel's normal)
        if xb-xa <= 0.:
            self.beta = numpy.arccos((yb-ya)/self.length)
        elif xb-xa > 0.:
            self.beta = math.pi + numpy.arccos(-(yb-ya)/self.length)
        
        # location of the panel
        if self.beta <= numpy.pi:
            self.loc = 'extrados'
        else:
            self.loc = 'intrados'
        
        self.sigma = 0.     # source strength
        self.vt = 0.        # tangential velocity
        self.cp = 0.        # pressure coefficient
```

```python
def define_panels(x, y, N):
    """Discretizes the geometry into panels.
    
    Arguments
    ---------
    x, y -- Cartesian coordinates of the geometry (1D arrays).
    N - number of panels (default 40).
    
    Returns
    -------
    panels -- Numpy array of panels.
    """
    x_ends = x
    y_ends = y

    panels = numpy.empty(N, dtype=object)
    for i in xrange(N):
        panels[i] = Panel(x_ends[i], y_ends[i], x_ends[i+1], y_ends[i+1])
    
    return panels
```

```python
class Freestream:
    """Freestream conditions."""
    def __init__(self, u_inf, alpha):
        """Sets the freestream conditions.
        
        Arguments
        ---------
        u_inf -- Farfield speed (default 1.0).
        alpha -- Angle of attack in degrees (default 0.0).
        """
        self.u_inf = u_inf
        self.alpha = alpha*math.pi/180          # degrees --> radians
```

```python
def integral(x, y, panel, dxdz, dydz):
    """Evaluates the contribution of a panel at one point.
    
    Arguments
    ---------
    x, y -- Cartesian coordinates of the point.
    panel -- panel which contribution is evaluated.
    dxdz -- derivative of x in the z-direction.
    dydz -- derivative of y in the z-direction.
    
    Returns
    -------
    Integral over the panel of the influence at one point.
    """
    def func(s):
        return ( ((x - (panel.xa - numpy.sin(panel.beta)*s))*dxdz 
                  + (y - (panel.ya + numpy.cos(panel.beta)*s))*dydz)
                / ((x - (panel.xa - numpy.sin(panel.beta)*s))**2 
                   + (y - (panel.ya + numpy.cos(panel.beta)*s))**2) )
    return integrate.quad(lambda s:func(s), 0., panel.length)[0]
```

```python
def source_matrix(panels):
    """Builds the source matrix.
    
    Arguments
    ---------
    panels -- array of panels.
    
    Returns
    -------
    A -- NxN matrix (N is the number of panels).
    """
    N = len(panels)
    A = numpy.empty((N, N), dtype=float)
    numpy.fill_diagonal(A, 0.5)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i,j] = 0.5/numpy.pi*integral(p_i.xc, p_i.yc, 
                                              p_j, numpy.cos(p_i.beta), numpy.sin(p_i.beta))
    
    return A
```

```python
def vortex_matrix(panels):
    """Builds the vortex array.
    
    Arguments
    ---------
    panels - array of panels.
    
    Returns
    -------
    a -- 2D array (Nx2, N is the number of panels).
    """
    N = len(panels)
    A = numpy.empty((N,N), dtype=float)
    numpy.fill_diagonal(A, 0.0)

    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i,j] -= 0.5/numpy.pi*integral(p_i.xc, p_i.yc, 
                                             p_j, numpy.sin(p_i.beta), -numpy.cos(p_i.beta))
    return A
```

```python
def Integral_array(matrix):
    """Finds the integral of a matrix
    
    Arguments
    ---------
    matrix -- the matrix either vortex or source
    
    Returns
    --------
    B -- (2xN) matrix (N is length of matrix)
    """
    N = len(matrix[0])
    N2 = N/2
    S1 = numpy.zeros((N,N2), dtype=float)
    S2 = numpy.zeros((N,N2), dtype=float)
    for i in range(0,N):
        for h in range(0,N2):
            S1[i][h] = matrix[i][h]
            S2[i][h] = matrix[i][h+N/2]
    B = numpy.zeros([N,2], dtype=float)
    B1 = numpy.zeros([N,1], dtype=float)
    B2 = numpy.zeros([N,1], dtype=float)
    for i in range(0,N):
        for h in range(0,(N/2)):
            B1[i][0] = B1[i][0]+S1[i][h]
            B2[i][0] = B2[i][0]+S2[i][h]

    for i in range(0,N):
        B[i][0] = B1[i]
        B[i][1] = B2[i]
    
    return B
```

```python
def kutta_cond(matrix):
    """Calculates the kutta condish of a combined matrix
    then spits out 2 rows by N columns.
    
    Arguments
    ----------
    matrix -- combined matrix
    
    Returns
    --------
    K -- (2,N) Matrix where N is length of matrix.
    """
    NumCol = len(matrix[0])
    NumRow = len(matrix)
    N2 = NumRow/2
    
    K = numpy.zeros((2,NumCol), dtype=float)
    
    for i in xrange(0,NumCol):
        K[0][i] = matrix[0][i] + matrix[N2-1][i]
        K[1][i] = matrix[N2][i] + matrix[NumRow-1][i]
    
    return K   
    
```

```python
def build_matrix(panels):
    """Builds the matrix of the linear system.
    
    Arguments
    ---------
    panels -- array of panels.
    
    Returns
    -------
    A -- (N+1)x(N+1) matrix (N is the number of panels).
    """
    N = len(panels)
    M = numpy.zeros((N+2, N+2), dtype=float)
    L = numpy.zeros((N+2, N+2), dtype=float)
    Bt = numpy.zeros((N, N+2), dtype=float)
    At = numpy.zeros((N, N+2), dtype=float)
    A_Source_Matrix = numpy.zeros((N,N), dtype=float)
    B_Vortex_Matrix = numpy.zeros((N,N), dtype=float)
    SumA = numpy.zeros((N,2),dtype=float)
    SumB = numpy.zeros((N,2),dtype=float)
    KuttaAt = numpy.zeros((2,N+2),dtype=float)
    KuttaBt = numpy.zeros((2,N+2),dtype=float)
    
    #Build our source matrix A, then create our SigmaA columns.
    A_Source_Matrix = source_matrix(panels)
    SumA = Integral_array(A_Source_Matrix)
    SumA = -1*SumA
    
    #Build our Concatenated Bn matrixes
    B_Vortex_Matrix = vortex_matrix(panels)
    for i in range(0,N):
        for h in range(0,N):
            Bt[i][h] = B_Vortex_Matrix[i][h]
            Bt[i][N] = SumA[i][0]
            Bt[i][N+1] = SumA[i][1]
    
    #Take Kutta condition of Bt
    KuttaBt = kutta_cond(Bt)
    
    #Sum the B matrix
    SumB = Integral_array(B_Vortex_Matrix)
    
    #Build our Big Matrix M
    for i in range(0,N):
        for h in range(0,N):
            M[i][h] = A_Source_Matrix[i][h]
            M[i][N] = SumB[i][0]
            M[i][N+1] = SumB[i][1]
            
    for i in range(0,N+2):
        M[N][i] = KuttaBt[0][i]
        M[N+1][i] = KuttaBt[1][i]
    
    #Time to start over.  We then need to build
    #our second big matrix, so we repeat the process
    #that we have laid out.
    #Create our Concatenated An matrix
#     for i in range(0,N):
#         for h in range(0,N):
#             At[i][h] = A_Source_Matrix[i][h]
#             At[i][N] = SumB[i][0]
#             At[i][N+1] = SumB[i][1]
    
#     #Take the kutta condition
#     KuttaAt = kutta_cond(At)
    
#     #Build our Big Matrix N
#     for i in range(0,N):
#         for h in range(0,N):
#             L[i][h] = B_Vortex_Matrix[i][h]
#             L[i][N] = SumA[i][0]
#             L[i][N+1] = SumA[i][1]
    
#     for i in range(0,N+2):
#         L[N][i] = KuttaAt[0][i]
#         L[N+1][i] = KuttaAt[1][i]
    
    return M#, L
```

```python
def build_rhs(panels, freestream1, freestream2):
    """Builds the RHS of the linear system.
    
    Arguments
    ---------
    panels -- array of panels.
    freestream -- farfield conditions.
    
    Returns
    -------
    b -- 1D array ((N+2)x1, N is the number of panels).
    """
    N = len(panels)
    N1 = N/2
    b = numpy.empty((N+2,1),dtype=float)

#     for i, panel in enumerate(panels):
#         b[i] = - freestream1.u_inf * math.cos(freestream1.alpha - panel.beta)
        
    for i, panel in enumerate(panels):
        if i < N1:
            b[i][0] = - freestream1.u_inf * numpy.cos(freestream1.alpha - panel.beta)
        
        if i >= N1:
            b[i][0] = - freestream2.u_inf * numpy.cos(freestream2.alpha - panel.beta)
    
    b[N][0] = -freestream1.u_inf*( numpy.sin(freestream1.alpha-panels[0].beta)
                              +numpy.sin(freestream1.alpha-panels[N1-1].beta) )
    b[N+1][0] = -freestream2.u_inf*( numpy.sin(freestream2.alpha-panels[N1].beta)
                              +numpy.sin(freestream2.alpha-panels[N-1].beta) )
    
    return b
```

```python
def get_tangential_velocities(panels, freestreamMain, freestreamFlap, gammaMain, gammaFlap, tanMatrix):
    """Computes the tangential velocity on the surface.
    
    Arguments
    ---------
    panels -- array of panels.
    freestream -- farfield conditions.
    gamma1 -- circulation density of airfoil.
    gamma2 -- circulation density of flap.
    tanMatrix -- tangential matrix
    
    """
    N = len(panels)
    N2 = N/2
    
    for i in range(0,N):
        if i < N2:
            b = freestreamMain.u_inf * numpy.sin([freestreamMain.alpha - panel.beta for panel in panels])
        if i >= N2:
            b = freestreamFlap.u_inf * numpy.sin([freestreamFlap.alpha - panel.beta for panel in panels])
    
    
    #b2 = freestreamFlap.u_inf * numpy.sin([freestreamFlap.alpha - panel.beta for panel in panels])
    
    var = numpy.append([panel.sigma for panel in panels], gammaMain)
    var1 = numpy.append(var, gammaFlap)
    
    vt = numpy.dot(tanMatrix, var1) + b
    
    for i, panel in enumerate(panels):
        panel.vt = vt[i]
```

```python
def build_tangent(panels):
    '''
    Arguments
    ---------
    panels -- panels you want in
    
    Outputs
    --------
    Bt -- tangent matrix
    '''
    
    #Code Copied from my build matrix function above. We are building the
    #tangential matrix with these.
    Ntan = len(panels)
    Bt = numpy.zeros((Ntan, Ntan+2), dtype=float)
    A_Source_Matrix = numpy.zeros((Ntan,Ntan), dtype=float)
    B_Vortex_Matrix = numpy.zeros((Ntan,Ntan), dtype=float)
    SumA = numpy.zeros((Ntan,2),dtype=float)

    #Build our source matrix A, then create our SigmaA columns.
    A_Source_Matrix = source_matrix(panels)
    SumA = Integral_array(A_Source_Matrix)
    SumA = -1*SumA

    #Build our Concatenated Bn matrixes
    B_Vortex_Matrix = vortex_matrix(panels)
    for i in range(0,Ntan):
         for h in range(0,Ntan):
             Bt[i][h] = B_Vortex_Matrix[i][h]
             Bt[i][Ntan] = SumA[i][0]
             Bt[i][Ntan+1] = SumA[i][1]
                
    return Bt
```

```python
def get_pressure_coefficient(panels, freestream1, freestream2):
    """Computes the surface pressure coefficients.
    
    Arguments
    ---------
    panels -- array of panels.
    freestream -- farfield conditions.
    """
    N = len(panels)
    N2 = N/2
    
    if i in range(0,N):
        if i < N2:
            for panel in panels:
                panel.cp = 1.0 - (panel.vt/freestream1.u_inf)**2
        if i >= N2:
            for panel in panels:
                panel.cp = 1.0 - (panel.vt/freestream2.u_inf)**2
```

```python
def CP_Coeff(panels):
    '''
    Arguments
    ---------
    panels -- enter panels
    
    Returns
    --------
    Prints out the Lift and Drag
    '''
    
    Lift = 0
    Drag = 0
    
    for panel in panels:
        Lift = Lift-panel.cp*panel.length*numpy.sin(panel.beta)
        Drag = Drag+panel.cp*panel.length*numpy.cos(panel.beta)
        
    return Lift, Drag
```

```python
def rotatePanel(cx, cy, foilx, foily, alpha):
    '''
    Arguments
    ---------
    cx, cy -- center coordinates x,y of rotation
    foilx, foily -- which foil point to rotate
    alpha -- angle of attack to rotate
    
    Returns
    --------
    X, Y - rotated values
    '''
    
    alpha = alpha*(numpy.pi/180)
    
    rotx = foilx - cx
    roty = foily - cy
    
    foilx = rotx*numpy.cos(alpha) - roty*numpy.sin(alpha) + cx
    foily = rotx*numpy.sin(alpha) + roty*numpy.cos(alpha) + cy
    
    return foilx, foily
    
```
